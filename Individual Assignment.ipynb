```python
import pandas as pd
```

### Import CSV

```python
imdb = pd.read_csv("https://raw.githubusercontent.com/scalabretta/GroupProject--IMDB/master/Dataset/IMDB-database-update.csv", encoding='latin1')
```

### How many rows and columns reflected on dataset file IMDB-database-update.csv?

```python
print(imdb.shape)
```

### What are the top 3 movies with highest nett profit in 2013?

```python
imdb_2013 = imdb[imdb['title_year'] == 2013].sort_values(by='net_USD', ascending=False)
imdb_2013
```

### What are the top 3 movies with highest gross profit in 2009? 

```python
imdb_2009 = imdb[imdb['title_year'] == 2013].sort_values(by='gross_USD', ascending=False)
imdb_2009
```

### How many non-english movies are in the dataset? 

```python
imdb_noneng = imdb[imdb['language'] != 'English']
len(imdb_noneng)
```

### What are top 3 languages in the dataset aside English (based on number of movies)?

```python
sort_noneng = imdb_noneng['language'].value_counts()
sort_noneng.head(3)
```

### What are the top 3 countries with the highest nett profit in all dataset times?

```python
top_country = imdb.groupby('country').mean().sort_values(by='net_USD', ascending=False).head(3)
top_country[['net_USD']].astype(int)
```

### What are the average budget of all movies in dataset in 2013?

```python
mean_2013 = imdb_2013['budget_USD'].mean()
mean_2013
```

### What are the average budget of all movies in dataset in 2014?

```python
imdb_2014 = imdb[imdb['title_year'] == 2014]
mean_2014 = imdb_2014['budget_USD'].mean()
mean_2014
```
